# PyMB Example

### Import Module

```python
import sys
sys.path.append('../..')
import PyMB
from PyMB import magic  # enable %%PyMB cell magic
import numpy as np
```

### Define the model

$$ Y \sim \mathcal{N}(\hat{Y},\sigma) $$
$$ \hat{Y} = \alpha + B x $$

Based on [https://github.com/kaskr/adcomp/blob/master/tmb_examples/linreg.cpp](https://github.com/kaskr/adcomp/blob/master/tmb_examples/linreg.cpp)

See [the TMB tutorial](https://github.com/kaskr/adcomp/wiki/Tutorial#writing-the-c-function) for more information on writing custom models

```python
%%PyMB LinearRegression
// DATA
    DATA_VECTOR(Y);
    DATA_VECTOR(x);

// PARAMETERS
    PARAMETER(alpha);
    PARAMETER(Beta);
    PARAMETER(logSigma);

// MODEL
    vector<Type> Y_hat = alpha + Beta*x;
    REPORT(Y_hat);
    Type nll = -sum(dnorm(Y, Y_hat, exp(logSigma), true));
    return nll;
```

### Simulate data

```python
LinearRegression.data = {
    'x': np.arange(10),
    'Y': np.random.normal(np.arange(10))
}
```

### Set initial parameter values

```python
LinearRegression.init = {
    'alpha': 0.,
    'Beta': 0.,
    'logSigma': 0.
}
```

### Fit the model

The model likelihood will be integrated wrt the random parameters. See [here](https://github.com/kaskr/adcomp/wiki/Tutorial#integrated-likelihood) for more information.

```python
LinearRegression.optimize(random=['alpha','Beta'])
```

### Extract fitted values

```python
print(LinearRegression.report('Y_hat'))
```

### Examine joint density

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
df = pd.DataFrame({ k: v['draws'][0] for k,v in LinearRegression.parameters.iteritems() })
g = sns.PairGrid(df, diag_sharey=False)
g.map_lower(sns.kdeplot, cmap='Blues_d')
g.map_upper(plt.scatter)
g.map_diag(sns.kdeplot, lw=3)
```
